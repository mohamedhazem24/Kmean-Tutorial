# The Whole Mechanism of K-means

```python
dist=np.zeros((df.shape[0],3))
xy=df[['x','y']].values
x=df['x'].values
y=df['y'].values
Centroids=np.zeros((3,2))
```

```python
for i in range(3):
    Centroids[i,0]=np.random.randint(450,700)
    Centroids[i,1]=np.random.randint(200,450)
plt.figure()
plt.scatter(df['x'],df['y'],c='blue')
plt.scatter(Centroids[:,0],Centroids[:,1],c="red")
plt.xlabel("Intialize Random Centroids")
plt.ylabel("First Step")
plt.show()
for i in range(8):
    dist[:,0]=np.sqrt(np.sum(np.square(xy-Centroids[0,:]),axis=1))
    dist[:,1]=np.sqrt(np.sum(np.square(xy-Centroids[1,:]),axis=1) )
    dist[:,2]=np.sqrt(np.sum(np.square(xy-Centroids[2,:]),axis=1))
    cluster=np.argmin(dist,axis=1)
    plt.figure(figsize=(20,5))
    plt.subplot(1,2,1)
    plt.ylabel("iteration no."+str(i+1))
    plt.scatter(x,y,c=cluster)
    plt.scatter(Centroids[:,0],Centroids[:,1],c='red')
    plt.xlabel("Map points to its Cluster")
    Centroids[0,0]=np.mean(x[cluster==0])
    Centroids[0,1]=np.mean(y[cluster==0])
    Centroids[1,0]=np.mean(x[cluster==1])
    Centroids[1,1]=np.mean(y[cluster==1])
    Centroids[2,0]=np.mean(x[cluster==2])
    Centroids[2,1]=np.mean(y[cluster==2])
    plt.subplot(1,2,2)    
    plt.scatter(x,y)
    plt.scatter(Centroids[:,0],Centroids[:,1],c='red')
    plt.xlabel("Update Centroids")
    plt.show()
    
```

# Mecahanism of K means (Step by Step)

## Step 1 : Generate Random Points

```python
np.random.seed(13)
Sample_Centroids=np.zeros((3,2))
for i in range(3):
    Sample_Centroids[i,0]=np.random.randint(450,700)
    Sample_Centroids[i,1]=np.random.randint(200,450)
plt.figure()
plt.scatter(df['x'],df['y'],c='blue')
plt.scatter(Sample_Centroids[:,0],Sample_Centroids[:,1],c="red")
plt.xlabel("Intialize Random Centroids")
plt.ylabel("First Step")
plt.show()
```

## Step 2 : Map each point to its cluster

### Step 2.A

We will Calculate The eculidean distance between points and each centroid to get to know which centroid is the nearest (have the shortest distance)


```python
dfCent=pd.DataFrame(Sample_Centroids,columns=["X","Y"],index=["C0 [Purple]","C1 [Yellow]","C2 [Green]"])
dfCent
```

```python
distances=np.zeros((10,3))
samples=np.zeros((10,2))

for i in range(10):
    sample_no=np.random.randint(0,df.shape[0])
    xpoint=x[sample_no]
    ypoint=y[sample_no]
    XCentroids=Sample_Centroids[:,0]
    YCentroids=Sample_Centroids[:,1]
    distances[i,0]=np.sqrt((xpoint-XCentroids[0])**2+(ypoint-YCentroids[0])**2)
    distances[i,1]=np.sqrt((xpoint-XCentroids[1])**2+(ypoint-YCentroids[1])**2)
    distances[i,2]=np.sqrt((xpoint-XCentroids[2])**2+(ypoint-YCentroids[2])**2)
    samples[i,0]=xpoint
    samples[i,1]=ypoint
    dist=pd.DataFrame(np.concatenate(([samples[i,:]],[distances[i,:]]),axis=1),columns=["X","Y","C0 [Purple]","C1 [Yellow]","C2 [Green]"],index=[i])
    display(dist)
    for j in range(3):
        print('Distance between point {} and Centroid {} = √ ({}-{})^2 + ({}-{})^2 = {}'.format(i,j,str(np.round(xpoint)),str(int(XCentroids[j])),str(np.round(ypoint)),np.round(YCentroids[j]),str(np.round(distances[i,j]))))
    display(dfCent)
    plt.figure()
    plt.title("Cluster 0 : Purple || Cluster 1 : Yellow || Cluster 2 : Green")
    plt.scatter(Sample_Centroids[:,0],Sample_Centroids[:,1],c=['purple','yellow','Green'])
    plt.scatter(x[sample_no],y[sample_no],c='blue')
    plt.plot([x[sample_no],Sample_Centroids[0,0],x[sample_no],Sample_Centroids[1,0],x[sample_no],Sample_Centroids[2,0]],[y[sample_no],Sample_Centroids[0,1],y[sample_no],Sample_Centroids[1,1],y[sample_no],Sample_Centroids[2,1]],linestyle='dashed')
    plt.show()
    print("------------------------------------------------------------------------------------------------------------------")
```

### Step 2.B

For each point we will choose the nearst centroid which has shortest distance with it 


```python
distdf=pd.DataFrame(np.concatenate((samples,distances),axis=1),columns=["X","Y","C0 [Purple]","C1 [Yellow]","C2 [Green]"])
distdf['Cluster']=distdf.columns[np.argmin(distances,axis=1)+2]
```

```python
distdf
```

```python
color_dict={"C2 [Green]":"green","C1 [Yellow]":"yellow","C0 [Purple]":"purple"}
for i in range(10):
    row=distdf.iloc[i]
    display(pd.DataFrame(row).T)
    plt.figure()
    plt.title("Cluster 0 : Purple || Cluster 1 : Yellow || Cluster 2 : Green")
    plt.scatter(Sample_Centroids[:,0],Sample_Centroids[:,1],c=['purple','yellow','Green'])
    plt.scatter(row['X'],row['Y'],c=color_dict[row['Cluster']])
    mxdist=np.sum([row['C0 [Purple]'],row['C1 [Yellow]'],row['C2 [Green]']])
    plt.plot([row['X'],Sample_Centroids[0,0]],[row['Y'],Sample_Centroids[0,1]],linestyle='dashed',color="blue",alpha=2-row['C0 [Purple]']/mxdist)
    plt.plot([row['X'],Sample_Centroids[1,0]],[row['Y'],Sample_Centroids[1,1]],linestyle='dashed',color="blue",alpha=2-row['C1 [Yellow]']/mxdist)
    plt.plot([row['X'],Sample_Centroids[2,0]],[row['Y'],Sample_Centroids[2,1]],linestyle='dashed',color="blue",alpha=2-row['C2 [Green]']/mxdist)
    plt.show()
    print("------------------------------------------------------------------------------------------------------------------")

    
```

## Step 3 : Update Our Old Centroid with new Centroids

We will update The centroids using these Fromulas



```python
plt.figure()
plt.scatter(distdf['X'],distdf['Y'],c=distdf['Cluster'].replace(color_dict))
plt.scatter(Sample_Centroids[:,0],Sample_Centroids[:,1],c="red")
plt.xlabel("Intialize Random Centroids")
plt.ylabel("First Step")
plt.show()
```

```python
Cluster0=distdf[distdf['Cluster'] =='C0 [Purple]']
sumstr=""
display(Cluster0)
for idx,i in enumerate(Cluster0['X']):
    if idx==Cluster0.shape[0]-1:
        sumstr=sumstr+str(np.round(i,2))
        continue
    sumstr=sumstr+str(np.round(i,2))+" + "
print("X of New Centroid = ({}) / {} = {}".format(sumstr,Cluster0.shape[0],np.mean(Cluster0['X']))) 
sumstr=""
for idx,i in enumerate(Cluster0['Y']):
    if idx==Cluster0.shape[0]-1:
        sumstr=sumstr+str(np.round(i,2))
        continue
    sumstr=sumstr+str(np.round(i,2))+" + "
print("Y of New Centroid = ({}) / {} = {}".format(sumstr,Cluster0.shape[0],np.mean(Cluster0['Y']))) 
print("Old Centroid ({},{})".format(Sample_Centroids[0,0],Sample_Centroids[0,1]))
print("New Centroid ({},{})".format(np.mean(Cluster0['X']),np.mean(Cluster0['Y'])))
```

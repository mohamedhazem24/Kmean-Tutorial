```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data= np.array([[1, 1], [1.5, 2], [3, 4], [5, 7],[3.5, 5], [4.5, 5], [3.5, 4.5]])
data
```

```python
dist = np.zeros((data.shape[0], 2),dtype=float)
print(dist)
xy = data
x = data[:, 0]
y = data[:, 1]
Centroids = np.array([[1, 1], [5, 7]],dtype=float)
```

```python
for i in range(2):
    dist[:, 0] = np.sqrt(np.sum(np.square(xy-Centroids[0, :]), axis=1))
    dist[:, 1] = np.sqrt(np.sum(np.square(xy-Centroids[1, :]), axis=1))
    cluster = np.argmin(dist, axis=1)
    df=pd.DataFrame(np.concatenate((xy,dist, np.expand_dims(cluster, axis=1)), axis=1),columns=["x","y",'C1',"C2","Cluster"])
    print(df)
    plt.figure(figsize=(20, 5))
    plt.subplot(1, 2, 1)
    plt.ylabel("iteration no."+str(i+1))
    plt.scatter(x, y, c=cluster)
    plt.scatter(Centroids[:, 0], Centroids[:, 1], c='red')
    plt.xlabel("Map points to its Cluster")
    print("Old Centroids\n",Centroids)
    Centroids[0, 0] = np.mean(x[cluster == 0])
    Centroids[0, 1] = np.mean(y[cluster == 0])
    Centroids[1, 0] = np.mean(x[cluster == 1])
    Centroids[1, 1] = np.mean(y[cluster == 1])
    print("New Centroids\n",Centroids) 
    plt.subplot(1, 2, 2)
    plt.scatter(x, y)
    plt.scatter(Centroids[:, 0], Centroids[:, 1], c='red')
    plt.xlabel("Update Centroids")
    plt.show()
```
